```python
# importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_selection import VarianceThreshold
from sklearn.preprocessing import LabelEncoder
```

```python
# loading datasets of both train and test 
train=pd.read_csv('/content/train.csv', index_col='ID')
test=pd.read_csv('/content/test.csv',index_col='ID')
```

#### unique values

```python
# checking for dtype for all columns
train.info()
```

```python
for col in train.columns:
  if(train[col].dtype != np.float64 and train[col].dtype != np.int64):
    unique_train=train[col].unique().tolist()
    unique_test=test[col].unique().tolist()

    for member in unique_test:
      if member not in unique_train:
        unique_train.append(member)

    #applying labelencoder
    le=LabelEncoder()  
    le.fit(list(train[col].values) + list(train[col].values))
    train[col] = le.transform(list(train[col].values))
    unique_value_train = train[col].unique().tolist()
    le.fit(list(test[col].values) + list(test[col].values))
    test[col] = le.transform(list(test[col].values))
    unique_value_test = test[col].unique().tolist()
```

###Variance Threshold

```python
# for any variance in column in train data=0 should be removed
var_thres_train=VarianceThreshold(threshold=0)
var_thres_train.fit(train)
```

```python
var_zero_train=[column for column in train.columns
if column not in train.columns[var_thres_train.get_support()]]
```

```python
# dropping columns with zero variance in axis=1
train.drop(var_zero_train, axis=1, inplace=True)
```

```python
train.head()
```

### checking for null values

```python
train.isnull().sum()
```

```python
# droping for zero variance for testdata
test.drop(var_zero_train,axis = 1,inplace = True)
test.head()
```

```python
#checking null for test data
test.isnull().sum()
```

### scaling

```python
X_train = train.iloc[:,1:]
y_train = train.iloc[:,0]

from sklearn.preprocessing import StandardScaler
scalling = StandardScaler()
X_train_scalled = scalling.fit_transform(X_train)
test_scalled = scalling.fit_transform(test)
```

### Dimensional Reduction

```python
# using PCA estimators
from sklearn.decomposition import PCA
```

```python
sklearn_PCA = PCA(n_components = 200)
```

```python
sklearn_PCA.fit(X_train_scalled)
x_train_transform = sklearn_PCA.transform(X_train_scalled)
print(x_train_transform.shape)
```

```python
#simillarly applying PCA for testdata
x_test_transform = sklearn_PCA.transform(test_scalled)
```

### XGBoost

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')
```

```python
grid_values = {'n_estimators': [74,75,76], 'learning_rate': [0.15,0.135,0.14], 'max_depth': [1,2,3,4]}
grid_xgb = GridSearchCV(xgb.XGBRegressor(), param_grid = grid_values, cv=10, scoring = 'r2',verbose = True)
grid_xgb.fit(X_train_scalled,y_train)
predict_xgb = grid_xgb.predict(test_scalled)

print('Mean score matrix: ', grid_xgb.cv_results_['mean_test_score'])
print('Grid best parameter (max. accuracy): ', grid_xgb.best_params_)
print('Grid best score (accuracy): ', grid_xgb.best_score_)
```

```python
final_predictions = pd.DataFrame()
final_predictions['id'] = test.index
final_predictions['y'] = pd.Series(predict_xgb)
final_predictions.to_csv('predictions.csv', index=False)
```
